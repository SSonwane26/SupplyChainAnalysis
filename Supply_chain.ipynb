```python
# Create the Customer Table
customer_table = supply[['Customer Id', 'Customer Fname', 'Customer Lname', 'Customer Segment','Customer City','Customer State', 'Customer Country']]
customer_table.head()
```

```python
# save in separate CSV file
customer_table.to_csv('customer_table.csv', index=False)
```

```python
# Create the Product Table
product_table = supply[['Product Card Id', 'Product Category Id', 'Product Name', 'Product Price']]
product_table.head()
```

```python
# save in separate CSV file
product_table.to_csv('product_table.csv', index=False)
```

```python
# Create the Category Table
category_table = supply[['Category Id', 'Category Name']]
category_table.head()
```

```python
# Create the Department Table
department_table = supply[['Department Id', 'Department Name', 'Latitude', 'Longitude']]
department_table.head()
```

```python
# save in separate CSV file
category_table.to_csv('category_table.csv', index=False)
department_table.to_csv('department_table.csv', index=False)
```

```python
# Create the Order Table
order_table = supply[['Order Id', 'Order Region', 'Order State']]
order_table.head()
```

```python
# save in separate CSV file
order_table.to_csv('order_table.csv', index=False)
```

```python
# Create the Order Item Table
order_item_table = supply[['Order Item Id', 'Order Item Cardprod Id', 'Order Item Product Price', 'Order Item Discount Rate']]
order_item_table.head()
```

```python
# save in separate CSV file
order_item_table.to_csv('order_item_table.csv', index=False)
```
